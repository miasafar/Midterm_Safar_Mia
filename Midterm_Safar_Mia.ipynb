```python
import pandas as pd
df= pd.read_csv('amazon-orders.csv')
df.head()
```

```python
df = df.fillna(0)
df.head()
```

```python
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
#Sum of money spent on Amazon
df["Total Charged"].sum()
```

```python
#Mean of money spent on Amazon
df["Total Charged"].mean()
```

```python
#Median of money spent on Amazon 
df["Total Charged"].median()
```

```python
#Maximum of money spent on Amazon
df["Total Charged"].max()
```

```python
#Minimum of money spent on Amazon
df["Total Charged"].min()
```

```python
#Sales tax 
df["Tax Charged"] = df["Tax Charged"].str.replace('$','').astype(float)
df.head()
```

```python
#Total money spent on taxes
df["Tax Charged"].sum()
```

```python
#Overall tax rate
df["Tax Charged"].sum() / df["Total Charged"].sum()
```

```python
df['Order Date'] = pd.to_datetime(df["Order Date"])
df.head()
```

```python
#How much spent on each day
df.plot.bar(x='Order Date', y="Total Charged", rot=90, figsize=(20,10))
```

```python
#Series of Daily Orders
daily_orders = df.groupby("Order Date").sum()["Total Charged"]
daily_orders.head()
```

```python
#Bar Graph of Daily Orders
daily_orders.plot.bar(figsize=(20,10))
```

```python
df['Order Date'] = pd.to_datetime(df["Order Date"])
df.head()
df.plot.pie(x='Order Date', y="Total Charged", rot=90, figsize=(20,10))
```

```python
df['Order Date'] = pd.to_datetime(df["Order Date"])
df.head()
df.plot.line(x='Order Date', y="Total Charged", rot=90, figsize=(20,10))
```
